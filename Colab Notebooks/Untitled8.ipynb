```
X = data2.iloc[:,data2.columns!='m13']
```

```
#without scaling
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.3,random_state=0)
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
classifier = RandomForestClassifier(random_state=42,max_depth=6,min_samples_split=2,n_estimators=200,n_jobs=-1,min_samples_leaf=2,criterion="entropy",class_weight={0:0.25,1:1})
```

```
classifier.fit(X_train,y_train)
```

```
y_predrd= classifier.predict(X_test)
```

```
from sklearn.metrics import f1_score
```

```
print("f1_score is:",f1_score(y_predrd,y_test))
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
classifier = DecisionTreeClassifier(random_state=0)
```

```
classifier.fit(X_train,y_train)
```

```
y_preddt= classifier.predict(X_test)
```

```
print("f1_score is:",f1_score(y_preddt,y_test))
```
